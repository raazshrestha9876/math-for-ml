| üîπ Point       | üîç Description                                               |
| -------------- | ------------------------------------------------------------- |
| **Definition** | A system to specify positions in space using ordered numbers. |
| **2D Point**   | Written as  (x, y) : X = horizontal, Y = vertical             |
| **3D Point**   | Written as  (x, y, z) : adds depth (Z = forward/back)         |
| **Used in**    | Geometry, physics, graphics, and machine learning             |
| **Real-world** | Maps (2D), video games (3D), plotting data                    |



üìò Simple Example
Imagine a drone flying in a 3D room:

It starts at (0, 0, 0)

Moves to (3, 4, 5)

So, it's:

3 units forward (x)

4 units up (y)

5 units into the room (z)

In 2D, (3, 4) means walking 3 steps right and 4 steps forward.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Problem: 2D and 3D cartesian points

Problem 1: 2D Plot

```python
plt.figure(figsize=(6, 6))
plt.plot(3, 4, 'ro')
plt.text(3, 4.2, '(3, 4)', fontsize=12)
plt.axhline(0, color="black", linewidth=0.5)
plt.axvline(0, color="black", linewidth=0.5)
plt.grid(True)
plt.title("2D Catesian Coordinate: (3, 4)")
plt.xlim(0, 6)
plt.ylim(0, 6)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

Problem 2: 3D PLOT

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(3, 4, 5, color='blue', s=50)
ax.text(3, 4, 5.5, '(3, 4, 5)', fontsize=12)
ax.set_xlim([0, 6])
ax.set_ylim([0, 6])
ax.set_zlim([0, 6])
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis (depth)')
ax.set_title('3D Cartesian Coordinate: (3, 4, 5)')
plt.show()
```
