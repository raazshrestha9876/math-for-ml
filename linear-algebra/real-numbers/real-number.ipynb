

✅ What are Real Numbers?

Real numbers (R) are all the values that can be found on the number line.

Components of Real Numbers:

| Type               | Example           | Definition                                    |
| ------------------ | ----------------- | --------------------------------------------- |
| Natural Numbers    | 1, 2, 3, ...      | Counting numbers                              |
| Whole Numbers      | 0, 1, 2, 3, ...   | Natural numbers + 0                           |
| Integers           | ..., -2, -1, 0, 1 | Whole numbers + negatives                     |
| Rational Numbers   | 1/2, 0.25, -4/5   | Can be written as p/q where q ≠ 0 |
| Irrational Numbers | π, sqrt(2)        | Cannot be written as a simple fraction        |
| Real Numbers       | All above         | Any point on the number line                  |

Real Number Formula Examples

1. Rational Number:
 
   3/4 =0.75(Rational)

2. Irrational Number:

   sqrt(2)=1.4142

3. Real-valued Function:

   f(x) = x^2 (Real input -> Real output )
   
Analogy

Imagine a long road:
Integer markers are like milestones (…-2, -1, 0, 1, 2…).
Fractions/decimals are like half-steps or precise coordinates.
Irrational points are places you can locate, but you can’t exactly measure with a ruler (like 𝜋).
Everything on this road is a real number.

```python
import numpy as np

values = [3, -7, 0.75, np.sqrt(2), np.pi, -3.14, "abc"]

def is_real_number(val):
    try:
        return isinstance(val, (int, float, np.float64)) and not np.isnan(val)
    except:
        return False

real_values = [val for val in values if is_real_number(val)]
print("Real numbers: ", real_values)
```

Problem 2: Visualize a Real-valued Function

formula:  

        f(x) = x^2

1. x ∈ [−10,10]
2. f(x) ∈ R (because x^2 is always real)


```python
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 400)
y = x ** 2

plt.plot(x, y, color='blue', label='y = x^2')
plt.xlabel("x (Real numbers)")
plt.ylabel("y")
plt.title("Real number function: y = x^2 ")
plt.grid(True)
plt.legend()
```
